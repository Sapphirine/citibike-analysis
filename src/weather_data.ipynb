```python
import datetime
from goose import Goose
import pickle
```

Get Daily temperature, precipation and snow data from www.almanac.com

```python
g = Goose()
temperature = []
prcpn = []
snow = []
for i in range(365):
    d = datetime.datetime(2015, 1, 1) + datetime.timedelta(i)
    date_str = d.strftime('%Y-%m-%d')
    url = 'https://www.almanac.com/weather/history/zipcode/10027/' + date_str
    article = g.extract(url)
    elem = article.raw_doc.find_class('weatherhistory_results_datavalue temp')[0]
    mean_temp = elem.getchildren()[1].text_content()
    print mean_temp[:4]
    temperature.append(mean_temp[:4])
    elem = article.raw_doc.find_class('weatherhistory_results_datavalue prcp')[0]
    mean_prcpn = elem.getchildren()[1].text_content()
    prcpn.append(mean_prcpn)
    elem = article.raw_doc.find_class('weatherhistory_results_datavalue sndp')[0]
    mean_snow = elem.getchildren()[1].text_content()
    snow.append(mean_snow)
```

```python
weather_data = (temperature, prcpn, snow)
```

```python
pickle.dump(weather_data, open('../data/weather_data.p', 'wb'))
```
